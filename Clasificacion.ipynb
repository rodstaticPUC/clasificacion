# Lirios: Clasificacion

- Datos: numero de muestras 150 (3 clases con 50 c/u)
- numero de atributos: 4 (longitud sepalo, anchura sepalo, longitud petalo y anchura petalo).
- Clase objetivo: lirio-setosa, lirio-versicolor, lirio-virginosa
- No hay datos faltantes 

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
target_names = iris.target_names
feature_names = iris.feature_names
print(target_names)
print(feature_names)
#print(iris.DESCR)
print(iris.data.shape) # dimensiones
```

```python
SEPAL_LENGTH = 0
fig, axs = plt.subplots(1,3, figsize=(15,4))

for ax, feature in zip(axs,[1,2,3]):
    for target, color in zip(range(3),['r','k','c']):
        ax.scatter(iris.data[iris.target==target,feature], 
               iris.data[iris.target==target,SEPAL_LENGTH],
                  color=color, marker="d", label=iris.target_names[target])
        ax.set_xlabel(iris.feature_names[feature], fontsize=15) # nombres eje x
        ax.set_ylabel(iris.feature_names[SEPAL_LENGTH], fontsize=15) # nombres eje y
        ax.legend(prop={'size':10}) #agregando leyenda
        ax.grid(True) # sumamos grid
```
